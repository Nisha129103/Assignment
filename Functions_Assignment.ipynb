#Q1.What is the difference between a function and a method in Python?
#Ans. In Python, the terms "function" and "method" are often used interchangeably, but there is a subtle difference:

Function:

- A standalone block of code that can be called independently.
- Not bound to any specific class or object.
- Can be defined inside or outside a class.
- Does not have access to the instance's attributes (self).

Method:

- A function that is defined inside a class and is bound to the instance of that class.
- Has access to the instance's attributes (self).
- Can modify the state of the instance.
- Is called on an instance of the class, not on the class itself.

In summary:

- Functions are standalone, while methods are part of a class.
- Functions don't have access to instance attributes, while methods do.

Here's an example:







````python
# Function
def greet(name):
    print(f"Hello, {name}!")
````

````python
greet("name")
````

````python
# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, {self.name}!")
````

````python
greet("self")
````

````python
# Calling the function
greet("John")
````

````python
# Calling the method
person = Person("Jane")
person.greet()
````

In this example, greet is a function, while greet inside the Person class is a method.

#Q6. Explain the concept of generators in Python and how they are defined.
#Ans. In Python, a generator is a special type of function that can be used to generate a sequence of values, but instead of building the entire sequence in memory, it generates values on-the-fly as they are needed.

A generator is defined using a function that contains a yield statement. When a generator function is called, it returns an iterator object, which can be used to iterate over the sequence of values.

Here's an example of a simple generator function:


````python
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1
````

In this example, the infinite_sequence function uses a while loop to generate an infinite sequence of numbers, starting from 0. The yield statement is used to produce each value in the sequence.

To use the generator, you can call the function and iterate over the resulting iterator object:


````python
seq = infinite_sequence()
for _ in range(10):
    print(next(seq))
````

Generators are useful when working with large datasets or infinite sequences, as they allow you to process values one at a time, without having to load the entire sequence into memory.

Some key benefits of generators include:

- Memory efficiency: Generators use significantly less memory than storing the entire sequence in a list or other data structure.
- Flexibility: Generators can be used to implement complex sequences and algorithms.
- Lazy evaluation: Generators only compute values as they are needed, which can improve performance.

#Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
#Ans. map(), reduce(), and filter() are three fundamental functions in Python's functools module that help with data processing and functional programming. Here's a brief overview of each:

1. map(function, iterable):
    - Applies a given function to each element of an iterable (like a list or tuple).
    - Returns a new iterable with the results.
    - Purpose: Transform data, apply a function to each element.

Example:numbers = [1, 2, 3]; doubled_numbers = list(map(lambda x: x*2, numbers))

1. reduce(function, iterable):
    - Applies a given function to the first two elements of an iterable, then to the result and the next element, and so on.
    - Returns a single output value.
    - Purpose: Aggregate data, reduce an iterable to a single value.

Example: import functools; numbers = [1, 2, 3]; sum = functools.reduce(lambda x, y: x+y, numbers)

1. filter(function, iterable):
    - Applies a given function to each element of an iterable and returns a new iterable with only the elements for which the function returns True.
    - Purpose: Select specific elements, filter out unwanted data.

Example: numbers = [1, 2, 3, 4, 5]; even_numbers = list(filter(lambda x: x%2 == 0, numbers))

In summary:

- map() transforms data by applying a function to each element.
- reduce() aggregates data by applying a function to all elements, reducing them to a single value.
- filter() selects specific elements by applying a function to each element and returning only those for which the function returns True.

These functions can be combined and used with lambda functions, making them powerful tools for data processing and functional programming in Python!

#Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
#Ans. Even Number Sum Function

Here's a Python function that calculates the sum of all even numbers in a given list:




````python
def sum_even_numbers(numbers):
    """
    Returns the sum of all even numbers in the input list.

    Args:
        numbers (list): A list of integers.

    Returns:
        int: The sum of all even numbers in the list.
    """
    even_sum = sum(num for num in numbers if num % 2 == 0)
    return even_sum
````

````python
# Test with a list containing only odd numbers
numbers = [1, 3, 5, 7]
result = sum_even_numbers(numbers)
print(result)
````

#Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

#Ans. Here's a Python function to check if a number is prime:


````python
def is_prime(n):
    """
    Checks if a number is prime.

    Args:
        n (int): The number to check.

    Returns:
        bool: True if prime, False otherwise.
    """
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
````

#Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

#Ans. Here's a Python iterator class that generates the Fibonacci sequence:


````python
class FibonacciIterator:
    """
    Iterator class to generate Fibonacci sequence up to a specified number of terms.
    """

    def __init__(self, max_terms):
        """
        Initialize the iterator.

        Args:
            max_terms (int): Maximum number of terms in the Fibonacci sequence.
        """
        self.max_terms = max_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        """
        Return the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Return the next number in the Fibonacci sequence.
        """
        if self.current_term < self.max_terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.current_term += 1
            return result
        else:
            raise StopIteration
````

````python
# Example usage
if __name__ == "__main__":
    max_terms = 10
    fib_iterator = FibonacciIterator(max_terms)

    print(f"Fibonacci sequence up to {max_terms} terms:")
    for term in fib_iterator:
        print(term)
````

Explanation


This `FibonacciIterator` class implements the iterator protocol by defining `__iter__` and `__next__` methods.


*   `__init__`: Initializes the iterator with the maximum number of terms.
*   `__iter__`: Returns the iterator object itself.
*   `__next__`: Returns the next number in the Fibonacci sequence.


To use the iterator, create an instance of `FibonacciIterator` with the desired number of terms and iterate over it using a `for` loop.

#Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

#Ans. Here's a Python generator function that yields the powers of 2:

````python
def powers_of_two(max_exponent):
    """
    Generator function to produce powers of 2 up to a given exponent.

    Args:
        max_exponent (int): Maximum exponent.

    Yields:
        int: Power of 2.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
````

````python
# Example usage
if __name__ == "__main__":
    max_exponent = 5
    print(f"Powers of 2 up to exponent {max_exponent}:")
    for power in powers_of_two(max_exponent):
        print(power)
````

Explanation


This powers_of_two generator function:


- Takes max_exponent as input.
- Iterates over the range from 0 to max_exponent (inclusive).
- Yields the power of 2 for each exponent using 2 ** exponent.

#Q7. Implement a generator function that reads a file line by line and yields each line as a string.

#Ans. Here's a Python generator function that reads a file line by line:


````python
def read_file_lines(file_path):
    """
    Generator function to read a file line by line.

    Args:
        file_path (str): Path to the file.

    Yields:
        str: Each line in the file.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
        return
````

````python
# Example usage
if __name__ == "__main__":
    file_path = 'example.txt'
    print(f"Reading file '{file_path}':")
    for line in read_file_lines(file_path):
        print(line)

````

Explanation


This read_file_lines generator function:


1. Opens the file in read mode ('r') using a with statement, ensuring automatic closure.
2. Iterates over each line in the file using a for loop.
3. Yields each line as a string, stripping leading/trailing whitespace using strip().
4. Handles FileNotFoundError exceptions, printing an error message.


Benefits


Using a generator function provides:


- Memory efficiency: Only one line is loaded into memory at a time.
- Flexibility: Use in loops, list comprehensions, or other iterations.


Alternative Implementation


You can also use the open function as a context manager with a for loop:

        return



This implementation uses the yield from syntax to delegate iteration to the generator expression.


Best Practices

- Strip whitespace from lines if necessary.

#Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

#Ans. Here's a Python program using map() to convert Celsius to Fahrenheit:

````python
def celsius_to_fahrenheit(celsius):
    """
    Convert Celsius to Fahrenheit.

    Args:
        celsius (float): Temperature in Celsius.

    Returns:
        float: Temperature in Fahrenheit.
    """
    return (celsius * 9/5) + 32
````

````python
# Sample list of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]
````

````python
# Use map() to convert Celsius to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
````

````python
# Print results
print("Celsius\tFahrenheit")
for celsius, fahrenheit in zip(celsius_temperatures, fahrenheit_temperatures):
    print(f"{celsius}\t\t{fahrenheit:.2f}")
````

Explanation


1. Define a function celsius_to_fahrenheit to perform the conversion.
2. Create a list celsius_temperatures containing temperatures in Celsius.
3. Use map() to apply the conversion function to each temperature.
4. Convert the map object to a list fahrenheit_temperatures.
5. Print the Celsius and Fahrenheit temperatures side-by-side.


Lambda Function Alternative


Replace the celsius_to_fahrenheit function with a lambda function:



fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))



This concise version achieves the same result.


Benefits of map()


- Applies a function to each element in an iterable.
- Returns a map object, which can be converted to a list or other iterable.
- Useful for data transformations and conversions.

#Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

#Ans. Here's a Python program using filter() to remove vowels:


````python
def is_consonant(char):
    """
    Check if a character is a consonant (not a vowel).

    Args:
        char (str): Single character.

    Returns:
        bool: True if consonant, False otherwise.
    """
    vowels = 'aeiouAEIOU'
    return char not in vowels
````

````python
# Sample string
input_string = "Hello, World!"
````

````python
# Use filter() to remove vowels
consonants = filter(is_consonant, input_string)
````

````python
# Join consonants into a string
result_string = ''.join(consonants)
````

````python
# Print results
print("Original String:", input_string)
print("String without Vowels:", result_string)
````

Explanation


1. Define a function is_consonant to check if a character is not a vowel.
2. Create a sample string input_string.
3. Use filter() to apply the is_consonant function to each character.
4. Join the filtered consonants into a string result_string.
5. Print the original and resulting strings.


Lambda Function Alternative


Replace the is_consonant function with a lambda function:



consonants = filter(lambda c: c not in 'aeiouAEIOU', input_string)



This concise version achieves the same result.


Benefits of filter()


- Applies a predicate function to each element in an iterable.
- Returns a filter object, which can be converted to a list or other iterable.
- Useful for data filtering and removal.


Tips


- Use is_consonant function for readability and reusability.
- Consider using regular expressions for more complex string operations.

#Q11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this.

#Ans. To solve this problem, you need to create a Python program that processes a list of orders and computes the total cost for each order. You will use lambda and map() to perform the required operations.

Here's a step-by-step breakdown and the corresponding Python program:

Define the Input Data:

1- Order numbers

2- Book titles & Authors

3- Quantities

4- Prices per item

Process the Data:

Compute the total cost for each order by multiplying the price per item by the quantity. Apply a conditional adjustment: if the computed cost is less than €100.00, increase it by €10.00.

Use lambda and map() to Transform the Data: Define a lambda function to compute the adjusted total cost. Use map() to apply this lambda function to each order.

Here’s the complete Python program to achieve this:

````python
# Define the input data
orders = [
    ("34587", 4, 40.95),
    ("98762", 5, 56.80),
    ("77226", 3, 32.95),
    ("88112", 3, 24.99)
]

# Define the lambda function to calculate adjusted total cost
adjusted_cost = lambda order: (
    order[0],  # Order number
    max(order[1] * order[2], 100) + (10 if order[1] * order[2] < 100 else 0)  # Adjusted cost
)

# Use map() to apply the lambda function to each order
result = list(map(adjusted_cost, orders))

# Print the result
print(result)
````
