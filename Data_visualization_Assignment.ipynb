#Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

#Ans. Here's a Python code snippet using NumPy, Pandas, Matplotlib, and Seaborn to generate a synthetic dataset and visualize the correlation matrix:



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic dataset with correlated features
n_samples = 1000
n_features = 5

# Correlation matrix
corr_matrix = np.array([
    [1.0, 0.8, 0.4, 0.2, 0.1],
    [0.8, 1.0, 0.7, 0.5, 0.3],
    [0.4, 0.7, 1.0, 0.9, 0.6],
    [0.2, 0.5, 0.9, 1.0, 0.8],
    [0.1, 0.3, 0.6, 0.8, 1.0]
])

# Generate random data
data = np.random.multivariate_normal(mean=[0, 0, 0, 0, 0], cov=corr_matrix, size=n_samples)

# Create DataFrame
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5'])

# Calculate correlation matrix
corr = df.corr()

# Create heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True)

# Set title
plt.title('Correlation Matrix')

# Show plot
plt.show()
```

This code:


1. Imports necessary libraries.
2. Sets a random seed for reproducibility.
3. Defines a correlation matrix.
4. Generates random data using `np.random.multivariate_normal`.
5. Creates a Pandas DataFrame.
6. Calculates the correlation matrix using `df.corr`.
7. Creates a heatmap using Seaborn's `heatmap`.
8. Adds title.


*Tips:*


- Customize colors using `cmap='viridis'`.
- Add mask for upper triangle using `mask=np.triu(np.ones_like(corr, dtype=bool))`.
- Save plot using `plt.savefig('correlation_matrix.png')`.


*Variations:*


- **Scatterplot matrix:** `sns.pairplot(df)`
- **Correlation network:** `sns.networkx(corr, threshold=0.5)`

#Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
!pip install bokeh
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
num_points = 50
data = {
    'x': np.random.rand(num_points) * 10,
    'y': np.random.rand(num_points) * 10,
    'sizes': np.random.randint(5, 20, num_points),
    'colors': np.random.choice(['red', 'blue', 'green'], num_points)
}
df = pd.DataFrame(data)

# Create a Bokeh plot
p = figure(title="Scatter Plot with Random Data", x_axis_label="x", y_axis_label="y")

# Add scatter glyphs with varying size and color
p.scatter(df['x'], df['y'], size=df['sizes'], color=df['colors'], alpha=0.7)

# Show the plot
show(p)
```

#Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
#Ans.
import numpy as np
from bokeh.plotting import figure, show

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create Bokeh plot
p = figure(title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           x_range=fruits)

# Add bar renderer to plot
p.vbar(x=fruits, top=counts, width=0.5, legend_label="Fruit Counts",
       line_color='white', fill_color="#CAB2D6")

# Show plot
show(p)
```

#Q4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

#Ans.


```python
import numpy as np
from bokeh.plotting import figure, show

# Generate random data
np.random.seed(0)
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create Bokeh plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency',
           x_range=[edges[0], edges[-1]])

# Add histogram renderer to plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="#CAB2D6", line_color="white")

# Show plot
show(p)
```

#Q5. Create a Bokeh heatmap using the provided dataset.

#Ans.

```python
!pip install bokeh
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
import numpy as np

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Reshape data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'values': data_heatmap.flatten()
}
source = ColumnDataSource(data)

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create Bokeh plot
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None, tools="")  # Remove toolbar and tools

# Add heatmap rectangles
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color={'field': 'values', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
